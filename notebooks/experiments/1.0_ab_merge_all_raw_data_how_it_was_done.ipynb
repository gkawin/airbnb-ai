### Concatenating all data files

```python
df_combined = pd.read_csv(csv_filepathes[0])
print(f'Processing {csv_filepathes[0]=}')
print(f'Combined shape {df_combined.shape=}\n')

for csv_file in csv_filepathes[1:]:
    print(f'Processing {csv_file=}')

    df = pd.read_csv(csv_file)
    print(f'Shape {df.shape=}')

    df_combined = pd.concat([df_combined, df])
    print(f'Combined shape {df_combined.shape=}\n')
```

### Reading description, amenities and house rules

```python
import json
```

```python


df_data = {
    "listing_id" : [],
    "checking_in_and_out_house_rule" : [],
    "during_your_stay_house_rule" : [],
    "bathroom_amenities" : [],
    "bedroom_and_laundry_amenities" : [],
    "entertainment_amenities" : [],
    "heating_and_cooling_amenities" : [],
    "privacy_and_safety_amenities" : [],
    "internet_and_office_amenities" : [],
    "kitchen_and_dining_amenities" : [],
    "location_features_amenities" : [],
    "outdoor_amenities" : [],
    "parking_and_facilities_amenities" : [],
    "services_amenities" : [],
    "place_description" : [],
    "the_space_description" : [],
    "guest_access_description" : [],
    "other_things_to_note_description" : [],
    "home_safety_amenities" : [],
    "not_included_amenities" : [],
    "scenic_views_amenities" : [],
    "during_your_stay_description" : [],
    "registration_number_description" : [],
    "family_amenities" : [],
    "before_you_leave_house_rule" : [],
}

def extract_listing_df(file_path):    
    with open(file_path, 'r') as file:
        for line in file:
            data = json.loads(line)
            
            for key in df_data.keys():
                df_data[key].append(data.get(key, "nan"))

    return df_data

# Replace 'your_file.jsonl' with the path to your actual .jsonl file
file_path = 'description_amenities_house_rules.jsonl'
json_listing_ids = extract_listing_df(file_path)
print(len(json_listing_ids.keys()), len(json_listing_ids['listing_id']))
```

```python
descr_amen_house_rules = pd.DataFrame(data=json_listing_ids)

print(descr_amen_house_rules.shape)

descr_amen_house_rules.head(20)
```

```python
print(df_raw.shape)

df_raw.head()
```

```python
# Perform the inner join on 'listing_id'
merged_df = pd.merge(df_raw, descr_amen_house_rules, right_on='listing_id', left_on='id', how='inner')

# Display the resulting DataFrame
print(merged_df.shape)
```

```python
pd.set_option('display.max_columns', 100)
```

```python
merged_df.drop("listing_id", axis=1, inplace=True, errors='ignore')
merged_df.rename(columns={'id': 'listing_id', 'userId': 'host_id'}, inplace=True, errors='ignore')
```

```python
import tqdm
import numpy as np
```

```python
# Function to convert DataFrame row to JSON serializable format
def convert_row_to_serializable(row):
    row_dict = row.to_dict()
    for key, value in row_dict.items():
        if isinstance(value, (pd.Timestamp, pd.Timedelta)):
            row_dict[key] = str(value)
        elif isinstance(value, (np.bool_, np.integer, np.floating, np.ndarray)):
            row_dict[key] = value.item()
    return row_dict
```

```python
output_file_path = 'raw_data_descriptions_amenities_house_rules.jsonl'

# Check if file exists and delete it
if os.path.exists(output_file_path):
    os.remove(output_file_path)
    print(f"{output_file_path} has been deleted.")
else:
    print(f"{output_file_path} does not exist.")

for _, row in merged_df.iterrows():
    # row_json = row.to_json()

    # Convert row to JSON serializable format
    serializable_row = convert_row_to_serializable(row)

    # Write the extracted data to the output file
    with open(output_file_path, 'a') as file:  # Open file in append mode
        file.write(json.dumps(serializable_row) + '\n')

print(f"{output_file_path} successfully saved.")

    
```
